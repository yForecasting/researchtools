## 4. Sentiment Analysis

### 4.1 Calculate sentiment score

```python
import pandas as pd
from textblob import TextBlob

# Read in the dataframe
df = pd.read_csv('literature_table.csv')

# For the titles: Get the second column of the dataframe
text = df.iloc[:, 5]

# Initialize an empty list to store the sentiment scores
sentiment_scores = []

# Iterate through the text and compute the sentiment score for each row
for t in text:
  sentiment_scores.append(TextBlob(t).sentiment.polarity)

# Add the sentiment scores to the dataframe as a new column
df['sentiment_score'] = sentiment_scores

# Print the dataframe
print(df.sentiment_score)
```

### 4.2 Show the results

```python
import matplotlib.pyplot as plt

# Extract the sentiment scores from the dataframe
sentiment_scores = df['sentiment_score']

# Create a histogram of the sentiment scores
plt.hist(sentiment_scores, bins=50)
plt.show()
```

```python
import seaborn as sns

# Create a box plot of the sentiment scores
sns.boxplot(sentiment_scores)
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Read in the dataframe
df = pd.read_csv('literature_table.csv', index_col=0)

# Get the abstract column of the dataframe
text = df.iloc[:, 5]

# Remove nan for my file
text[19]=" "
text[80] = " "

# Create a CountVectorizer object to convert the text to a matrix of token counts
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(text)

# Set the number of topics (change this value to adjust the number of topics)
n_topics = 10

# Create an LDA model
lda = LatentDirichletAllocation(n_components=n_topics)

# Fit the model to the data
lda.fit(X)

str_topics=[]
# Print the top words for each topic
for topic_idx, topic in enumerate(lda.components_):
  print("Topic %d:" % (topic_idx + 1))
  print(" ".join([vectorizer.get_feature_names()[i] for i in topic.argsort()[:-10 - 1:-1]]))
  str_topics.append(" ".join([vectorizer.get_feature_names()[i] for i in topic.argsort()[:-10 - 1:-1]])
```

## Abstract Summarisation

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

# Read in the dataframe
df = pd.read_csv('literature_table.csv', index_col=0)

# Get the second column of the dataframe
text = df.iloc[:, 5]

# Remove nan for my file
text[19]=" "
text[80] = " "

# Set the desired length of the summary (change this value to adjust the length)
summary_length = 100

# Create a TfidfVectorizer object to convert the text to a matrix of TF-IDF features
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(text)

# Create a TruncatedSVD object to perform dimensionality reduction
svd = TruncatedSVD(n_components=1)

# Fit the model to the data
svd.fit(X)

# Get the singular values and explained variance ratio
singular_values = svd.singular_values_
explained_variance_ratio = svd.explained_variance_ratio_

# Compute the length of the summary
summary_length = int(explained_variance_ratio * summary_length)
if summary_length < 7:
  summary_length=7

# Initialize an empty list to store the summaries
summaries = []

# Iterate through the text and compute the summary for each row
for i, t in enumerate(text):
  summary = " ".join(t.split()[:summary_length])
  summaries.append(summary)

# Add the summaries to the dataframe as a new column
df['summary'] = summaries

# Print the dataframe
print(df.summary)
```

If the explained variance ratio is very low, it means that the singular values are not explaining much of the variance in the data, and the summary will be short. This could be because the text is very homogeneous and does not contain much variation, or because the TruncatedSVD model is not performing well on the data.
