```python
from pybtex.database.input import bibtex
import pandas as pd

#open a bibtex file
parser = bibtex.Parser()
bibdata = parser.parse_file("literature.bib")
bibkeys=bibdata.entries.keys()
bibarray= []

for bib_id in bibdata.entries:
     b = bibdata.entries[bib_id].fields
     if b["year"] is None: 
        dyear='NA'
        print(dyear)
     else: 
        dyear=b["year"]

     if b.get("journal"):
       djournal=b["journal"]
     else: 
       djournal='NA'  

     if b.get("url"):
       durl=b["url"]
     else: 
       durl='NA'     

     if b.get("abstract"):
       dabstract=b["abstract"]
     else: 
       dabstract='NA'  

     d = {
            'bib_id' : bib_id,  # some formula for obtaining values
            'Title' : b["title"],
            'Year' : dyear,
            'Journal' : djournal,
            'URL' : durl,
            'Abstract' : dabstract
            }
     bibarray.append(d)

bibdataset = pd.DataFrame(bibarray)
```
